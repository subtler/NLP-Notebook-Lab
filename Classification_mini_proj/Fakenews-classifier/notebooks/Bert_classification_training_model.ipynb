## Load Tokenizer and Model (e.g., BERT for classification)

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from transformers import pipeline

#Load tokenizer and model

model_name = 'distilbert-base-uncased-finetuned-sst-2-english'
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

## Prepare Input and Run Prediction

```python
text = 'The government has announced a new initiative to boost employment.'

#tokenize input
inputs = tokenizer(text, return_tensors='pt')

#Forward pass
outputs = model(**inputs)

#Get predicted class
logits = outputs.logits
predicted_class_id = logits.argmax().item()

#Get label
labels = model.config.id2label
print('Predicted label:', labels[predicted_class_id])
```

```python
# Save
model.save_pretrained("my_model_directory")
tokenizer.save_pretrained("my_model_directory")

# Zip the correct folder
import shutil
shutil.make_archive("my_model_directory", 'zip', "my_model_directory")

# Download
from google.colab import files
files.download("my_model_directory.zip")
```
