## Step 2: Build Your First Chain

```python
from langchain_openai import ChatOpenAI  
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain


# Load the model
llm = ChatOpenAI(model_name="gpt-3.5-turbo",
                 temperature=0,
                 api_key='you_api_key'
                )

# Define a prompt template with a variable
prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners."
)

#create an LLM chain
chain= LLMChain(llm=llm, prompt=prompt)

#Run the chain with an input
response = chain.run('reinforement learning')
print("ðŸ§  Output:\n", response)
```

What you just built->

PromptTemplate:
Defines the input and structure of your prompt.

ChatOpenAI:
Wraps the model (could also be Cohere, HuggingFace, etc.)

LLMChain:
Glues model + prompt together into a callable pipeline

chain.run(...):
Sends input â†’ gets LLM output
